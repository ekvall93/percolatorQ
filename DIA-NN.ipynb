# Using Fragpipe's labling each replicate in each group -> One spectral library per replicate

## Run Diann with each spectral library, with its corresponding raw file.

```python
def runDiann(short_name, exp_name):
    for r_n in range(1,6):
        print("-------- Get quant for replicate {} in experiment group {}".format(r_n, exp_name))
        print("\n")
        path = '/home/ekvall/Desktop/sample_A_and_S/{}_{}'.format(short_name, r_n)
        lib = path + "/library.tsv"
        out = path + "/diann_out.tsv"
    
        file = "/sdd/proteomics/DATA/PXD014076/{}/mzML/{}_Proteome_DIA_RAW_{}0{}.mzML".format(exp_name, exp_name, short_name, r_n)
    
        run = " ".join([diann, "--f", file, "--lib", lib, "--out", out])
    
        !{run}
```

### Run for each replicate in the Aposymb group

```python
runDiann("A","Aposymb")
```

### Run for each replicate in the Symb group

```python
runDiann("S","Symb")
```

### Collect all diann output for each replicate

```python
def getQuantFram(short_name_list):
    data = list()
    for short_name in short_name_list:
        for r_n in range(1,6):
            path = '/home/ekvall/Desktop/sample_A_and_S/{}_{}'.format(short_name, r_n) + "/library.tsv"
            
            df = pd.read_csv(path, sep="\t")
            if short_name == "A":
                
                df["Run"] = ["Aposymb_Proteome_DIA_RAW_A0{}".format(r_n)] * df.shape[0]
                
            else:
                df["Run"] = ["Symb_Proteome_DIA_RAW_S0{}".format(r_n)] * df.shape[0]
            data.append(df)
    return pd.concat(data)
```

```python
df_one = getQuantFram(["A", "S"])
```

```python
df_one.sort_values(["ProteinId", "Run"]).tail(30).T
```

# Not using Fragpipe's labling for each replicate in each group -> One collective spectral library

## Run diann with the collective spectral library on all raw files.

```python
preSyntx = [diann, "--lib", "/home/ekvall/Desktop/sample_A_and_S_one_lib/library.tsv", "--out", "/home/ekvall/Desktop/sample_A_and_S_one_lib/diann_out.tsv"]
```

```python
def runAll(short_name_list, exp_name_list):
    files = list()
    pre_path = "/sdd/proteomics/DATA/PXD014076/"
    for short_name, exp_name in zip(short_name_list, exp_name_list):
        for r_n in range(1,6):
            file = pre_path + "{}/mzML/{}_Proteome_DIA_RAW_{}0{}.mzML".format(exp_name, exp_name, short_name, r_n)
            
            files += ["--f", file]
    return files
```

```python
files = runAll(["S", "A"],["Symb", "Aposymb"])
```

```python
syntax = " ".join(preSyntx + files)
```
